```python
df_2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz")

df_2000
```

```python
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
df_2001
```

## Concatenate dataframes

```python
df_combined = pd.concat([df_2000, df_2001])
```

```python
df_combined.head()
```

```python
df_combined.shape
```

```python
df_combined.info()
```

```python
revenue_filter = df_combined['revenue'] > 0
```

```python
budget_filter = df_combined['budget'] > 0
```

```python
df_filter = revenue_filter | budget_filter
```

```python
df_filter
```

```python
final_df=df_combined.loc[df_filter]
```

```python
final_df.shape
```

## Certification Categories

```python
final_df['certification'].value_counts()
```

## Average Revenue


```python
avg_revenue = final_df.groupby('certification')['revenue'].mean().round(1)
avg_revenue
```

## Average Budget

```python
avg_budget = final_df.groupby('certification')['budget'].mean().round(1)
avg_budget
```
